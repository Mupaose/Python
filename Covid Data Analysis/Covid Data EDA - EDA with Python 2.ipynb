### Overall Death Ratio

This is the ratio of reported deaths to reported cases and it calculated using the following formula:
    
* *death ratio* = new deaths / new cases
    

```python
death_rate = covid_df.new_deaths.sum() / covid_df.new_cases.sum()

print(death_ratio)
```

The figure provided is a decimal and we have to convert it to percentage for better undertstanding of the data

```python
percent_rate = death_rate * 100

print(percent_rate)
```

```python
#to the nearest 2 decimal places

print("The overall reported death rate is {:.2f}%.".format(percent_rate))
```
